```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from sklearn.impute import KNNImputer
from scipy import stats
```

```python
# Charger le dataset nettoyé
building_consumption = pd.read_csv('building_consumption_clean.csv', index_col=0)
print(building_consumption.shape)
print(building_consumption.columns.tolist())
```

```python
# Remplacer "Other" dans PrimaryPropertyType par LargestPropertyUseType
mask_other = building_consumption['PrimaryPropertyType'] == 'Other'
print(f"Nombre de bâtiments 'Other' : {mask_other.sum()}")

building_consumption.loc[mask_other, 'PrimaryPropertyType'] = \
    building_consumption.loc[mask_other, 'LargestPropertyUseType']

# Vérifier
print(f"\nNouvelles catégories dans PrimaryPropertyType :")
print(building_consumption['PrimaryPropertyType'].value_counts())
```

```python
# Garder uniquement les catégories avec plus de 10 occurrences
top_categories = building_consumption['PrimaryPropertyType'].value_counts()
top_categories = top_categories[top_categories >= 10].index

building_consumption['PrimaryPropertyType'] = building_consumption['PrimaryPropertyType'].apply(
    lambda x: x if x in top_categories else 'Other'
)

print(f"Nombre de catégories : {building_consumption['PrimaryPropertyType'].nunique()}")
building_consumption['PrimaryPropertyType'].value_counts()
```

```python
#Selection
from sklearn.model_selection import (
    train_test_split,
    GridSearchCV, 
    cross_validate,
    KFold
)
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error 
from sklearn.inspection import permutation_importance

#Preprocess
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler

#Modèles
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
# Supprimer les colonnes inutiles
building_consumption = building_consumption.drop(columns=['LargestPropertyUseType', 'BuildingType'])

```

```python
# Séparation X et y
X = building_consumption.drop(columns=['TotalGHGEmissions_log'])
y = building_consumption['TotalGHGEmissions_log']

# One-Hot Encoding
cat_cols = ['PrimaryPropertyType', 'Neighborhood']
enc = OneHotEncoder(sparse_output=False, drop='first', max_categories=15)
encoded = enc.fit_transform(X[cat_cols])
encoded_cols = enc.get_feature_names_out(cat_cols)
encoded_df = pd.DataFrame(encoded, columns=encoded_cols, index=X.index)
X = X.drop(columns=cat_cols)
X = pd.concat([X, encoded_df], axis=1)

print(f"Shape X : {X.shape}")
print(f"Catégorielles restantes : {X.select_dtypes(include=['object', 'category']).columns.tolist()}")
```

```python
def perform_cross_validation(
    X_train,
    y_train,
    model,
    cross_val_type,
    scoring_metrics: tuple,
    return_estimator=False,
    groups=None,
):
    scores = cross_validate(
        model,
        X_train,  
        y_train,
        cv=cross_val_type,
        return_train_score=True,
        return_estimator=return_estimator,
        scoring=scoring_metrics,
        groups=groups,
    )
    for metric in scoring_metrics:
        print(f"Average Train {metric} : {np.mean(scores['train_' + metric]):.4f}")
        print(f"Train {metric} Std : {np.std(scores['train_' + metric]):.4f}")
        print(f"Average Test {metric} : {np.mean(scores['test_' + metric]):.4f}")
        print(f"Test {metric} Std : {np.std(scores['test_' + metric]):.4f}")
        print("---")
    return scores
```

```python
# Séparation outliers
iqr = np.quantile(y, q=[0.25, 0.75])
limite_basse = iqr[0] - 1.5*(iqr[1] - iqr[0])
limite_haute = iqr[1] + 1.5*(iqr[1] - iqr[0])

mask_outliers = (y < limite_basse) | (y > limite_haute)
X_train = X[~mask_outliers]
y_train = y[~mask_outliers]

print(f"X_train : {X_train.shape}")

# Cross validation
cv = KFold(n_splits=5, shuffle=True, random_state=42)
scoring_metrics = ('r2', 'neg_mean_absolute_error', 'neg_root_mean_squared_error')

scores_lr = perform_cross_validation(
    X_train=X_train,
    y_train=y_train,
    model=LinearRegression(),
    cross_val_type=cv,
    scoring_metrics=scoring_metrics
)
```

```python
scores_rf = perform_cross_validation(
    X_train=X_train,
    y_train=y_train,
    model=RandomForestRegressor(random_state=42),
    cross_val_type=cv,
    scoring_metrics=scoring_metrics
)
```

```python
# Entraîner le modèle final sur tout X_train
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

# Permutation Importance
result = permutation_importance(
    model_lr, X_train, y_train, 
    n_repeats=10, 
    random_state=42,
    n_jobs=-1
)

# Afficher sous forme d'histogramme
feature_names = X_train.columns
sorted_idx = result.importances_mean.argsort()

plt.figure(figsize=(10, 12))
plt.barh(feature_names[sorted_idx], result.importances_mean[sorted_idx])
plt.xlabel("Permutation Importance")
plt.title("Feature Importance — Régression Linéaire")
plt.tight_layout()
plt.show()
```
